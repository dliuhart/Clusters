```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
import numpy as np
import pandas as pd


from sklearn.mixture import GaussianMixture
data = pd.read_csv("Ob1.csv");
rData = data[data.t ==0.1]
fitData = []
for n in range(0, len(rData['x'])):
 fitData.append([rData['x'].values[n],rData['y'].values[n],rData['z'].values[n]])

gm = GaussianMixture(n_components=12, random_state=0).fit(fitData)
gmfit = gm.predict(fitData)
xs = rData['x'].values
ys = rData['y'].values
zs = rData['z'].values
fig = plt.figure(figsize=(7,7))

ax = plt.axes(projection='3d')
ax.scatter3D(xs, ys, zs, c=gmfit, cmap='viridis')
#ax.set_size_inches(6, 6)
#ax.w_xaxis.set_ticklabels([])
#ax.w_yaxis.set_ticklabels([])
#ax.w_zaxis.set_ticklabels([])
ax.set_xlabel('x(mm)')
ax.set_ylabel('y(mm)')
ax.set_zlabel('z(mm)')
#ax.tick_params(labelbottom=True, labeltop=False, labelleft=True, labelright=True)
ax.dist = 12
ax.set_title('Clusters by using GMM', fontweight='bold')
plt.savefig('fig1a.eps', format='eps')
plt.show()
```
